```python
import cv2

# Lista de joints en el orden deseado
joints = [
    "pelvis",
    "spine",
    "spine.001",
    "neck",
    "face",
    "shoulder.L",
    "upper_arm.L",
    "forearm.L",
    "shoulder.R",
    "upper_arm.R",
    "forearm.R",
    "pelvis.L",
    "pelvis.R",
    "thigh.L",
    "shin.L",
    "foot.L",
    "heel.02.L",
    "thigh.R",
    "shin.R",
    "foot.R",
    "heel.02.R"
]

# Diccionario para almacenar las coordenadas de los joints
joint_coordinates = {}
jointDrawingCoordinates = {}

# Variables para controlar el modo de recopilación de datos
current_joint_index = 0  # Índice del joint actual
click_count = 0  # Contador de clics

# Función para manejar eventos de clic
def click_event(event, x, y, flags, params):
    global click_count, current_joint_index

    # Verifica si hay más joints por procesar
    if current_joint_index < len(joints):
        joint_name = joints[current_joint_index]

        if event == cv2.EVENT_LBUTTONDOWN:
            if click_count == 0:
                # Guarda las coordenadas en el diccionario
                joint_coordinates[joint_name] = (x, y)

                # Muestra las coordenadas en la consola
                print(f"{joint_name}: ({x}, {y})")

                # Muestra las coordenadas y el nombre del joint en la imagen
                font = cv2.FONT_HERSHEY_SIMPLEX
                cv2.putText(img, f"{joint_name}: {x},{y}", (x, y), font, 0.5, (255, 0, 0), 2)
                cv2.imshow('image', img)

                # Incrementa el contador de clics
                click_count += 1

            elif click_count == 1:
                # Recolecta información adicional en la segunda pulsación de clic (por ejemplo, cuadros de texto)

                # Muestra información adicional en la imagen
                # Esto es solo un ejemplo, debes agregar aquí tu lógica para recolectar la información adicional
                jointDrawingCoordinates[joint_name] = (x, y)
                font = cv2.FONT_HERSHEY_SIMPLEX
                cv2.putText(img, f"{joint_name} drawing coords: {x},{y}", (x, y), font, 0.5, (255, 0, 0), 2)
                cv2.imshow('image', img)


                # Reinicia el contador de clics para el siguiente joint
                click_count = 0

                # Incrementa el índice del joint actual
                current_joint_index += 1

# Función principal
if __name__ == "__main__":
    # Lee la imagen
    img = cv2.imread('./constantVariablesAllFilesSkeleton.png', 1)

    # Verifica si la imagen se cargó correctamente
    if img is None:
        print("Error: No se pudo cargar la imagen. Verifica la ruta del archivo y asegúrate de que el archivo existe.")
    else:
        # Muestra la imagen
        cv2.imshow('image', img)

        # Establece el manejador de eventos para clics del ratón
        cv2.setMouseCallback('image', click_event)

        # Espera a que se presione una tecla para salir
        cv2.waitKey(0)

        # Cierra la ventana
        cv2.destroyAllWindows()

        # Muestra el diccionario final con las coordenadas de los joints
        print("Coordenadas de los joints:")
        for joint, coords in joint_coordinates.items():
            print(f"{joint}: {coords}")
```
