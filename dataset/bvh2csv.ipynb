```python
import numpy as np
import os
np.set_printoptions(precision=6)

for root, dirs, files in os.walk("./csv"):
    for filename in files:
        if filename.endswith(".bvh"):  
            bvhFilePath = os.path.join(root, filename) 
            csvFilePath = bvhFilePath.replace(".bvh", ".csv")  
            jointInfo = dict()
            hierarchyNames = []
            csvToSave = open(csvFilePath, "w+")


            # Abre el archivo BVH y extrae información de los joints
            with open(bvhFilePath, "r") as bvhFile:
                actualJoint = ""
                initialJointPosition = []
                whatDegreesOfFreedom = []
                dof = 0
                for line in bvhFile:
                    splittedLine = line.split()
                    if "JOINT" in splittedLine or "ROOT" in splittedLine:
                        actualJoint = splittedLine[1]
                        hierarchyNames.append(actualJoint)
                    if "OFFSET" in splittedLine:
                        initialJointPosition = splittedLine[1:]
                    if "CHANNELS" in splittedLine:
                        dof = int(splittedLine[1])
                        whatDegreesOfFreedom = splittedLine[2:]
                        allInfo = {
                            "name": actualJoint,
                            "initialJointPosition": initialJointPosition,
                            "degreesOfFreedom": dof,
                            "whatDegreesOfFreedom": whatDegreesOfFreedom
                        }
                        jointInfo[actualJoint] = allInfo
                    if "Frame" in splittedLine:  # en la siguiente linea empieza frame a frame las posiciones y rotaciones
                        break

                #Escribir en el archivo csv, como primera fila los nombres de los joints que estan en el diccionario jointInfo
                for i, key in enumerate(jointInfo):
                    for dof in jointInfo[key]["whatDegreesOfFreedom"]:
                        csvToSave.write(key + "_" + dof + ",")

                
                csvToSave.write("\n")


                for dataLine in bvhFile:
                    splittedDataLine = dataLine.split()
                    #escribir las lineas de informacion de los frames en el archivo csv
                    nData = len(splittedDataLine)
                    for data in splittedDataLine:
                        csvToSave.write(data.strip() + ",")
                    csvToSave.write("\n")
                csvToSave.close()
                #cargar el archivo csv y comprobar el tamaño la columna de los nombres de los joints
                data = np.genfromtxt(csvFilePath, delimiter=",", skip_header=0)
                if not data.shape[1] == 112:
                    print("Error en el archivo: " + csvFilePath)
                    print("El archivo tiene " + str(data.shape[1]) + " columnas")
                    
                

```

```python
#cargar todos los csv y quitarles la ultima columna que existe (que esta vacia) y volver a guardarlos
import pandas as pd
for root, dirs, files in os.walk("./csv"):
    for filename in files:
        if not filename.endswith(".csv"):
            continue
        df = pd.read_csv(os.path.join(root, filename))
        df = df.iloc[:, :-1]
        df.to_csv(os.path.join(root, filename), index=False)
        print("File " + filename + " processed successfully!")
```
